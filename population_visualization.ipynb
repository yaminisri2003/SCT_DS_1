```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("population_data.csv", skiprows=4)
df.head()
```

```python
india_data = df[df["Country Name"] == "India"]
india_data
```

```python
years = india_data.columns[4:-1]
pop = india_data.loc[:, years].values.flatten()
years = [int(y) for y in years]
pop = [int(p)/1e9 for p in pop]

plt.figure(figsize=(10,5))
plt.plot(years, pop, marker='o', color='green')
plt.title("India's Population Over Time")
plt.xlabel("Year")
plt.ylabel("Population (Billions)")
plt.grid(True)
plt.show()
```
